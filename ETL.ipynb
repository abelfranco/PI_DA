```python
# Cambiar de 'Año' a 'Anio' desde el df1... hasta el df8
# Obtener una lista con los nombres de los dataframes
nombres_dataframes = [f'df{i}' for i in range(1, 9)]

# Iterar sobre los nombres de los dataframes
for nombre_df in nombres_dataframes:
    # Obtener el dataframe correspondiente
    df = globals()[nombre_df]
    
    # Cambiar el nombre de la columna "Año" a "Anio"
    df = df.rename(columns={"Año": "Anio"})
    
    # Asignar el dataframe modificado de vuelta a la variable global
    globals()[nombre_df] = df
```

```python
def reemplazar_vocales_con_tilde(dataframes):
    for i, df in enumerate(dataframes, 1):
        df.replace({'á': 'a', 'é': 'e', 'í': 'i', 'ó': 'o', 'ú': 'u'}, regex=True, inplace=True)
        print(f"Vocales con tilde reemplazadas en df{i}")

# Lista de dataframes
dataframes = [df1, df2, df3, df4, df5, df6, df7, df8, df9, df10, df11]

# Aplicar reemplazo en los dataframes
reemplazar_vocales_con_tilde(dataframes)
```

```python
def reemplazar_guión_por_nulos(dataframes):
    for i, df in enumerate(dataframes, 1):
        df.replace('--', np.nan, inplace=True)
        print(f"Valores '--' reemplazados por nulos en df{i}")

# Lista de dataframes
dataframes = [df1, df2, df3, df4, df5, df6, df7, df8, df9, df10, df11]

# Aplicar reemplazo en los dataframes
reemplazar_guión_por_nulos(dataframes)
```

```python
def reemplazar_coma_por_punto(dataframes):
    for i, df in enumerate(dataframes, 1):
        for columna in df.columns:
            if df[columna].dtype == 'object':
                df[columna] = df[columna].str.replace(',', '.')
        print(f"Valores ',' reemplazados por '.' en df{i}")

# Lista de dataframes
dataframes = [df1, df2, df3, df4, df5, df6, df7, df8, df9, df10, df11]

# Aplicar reemplazo en los dataframes
reemplazar_coma_por_punto(dataframes)
```

```python
import pandas as pd

# Lista para almacenar las columnas que deben ser convertidas a tipo float
def convert_to_float(df):
    columnas_convertir = []

    # Recorrer todas las columnas del dataframe
    for columna in df.columns:
        # Verificar si la columna contiene algún valor no nulo y es convertible a tipo float
        if pd.to_numeric(df[columna], errors='coerce').notnull().all():
            columnas_convertir.append(columna)

    # Convertir las columnas a tipo float
    df[columnas_convertir] = df[columnas_convertir].astype(float)

    return df

# Aplicar la conversión a tipo float para cada dataframe
df3 = convert_to_float(df3)
df4 = convert_to_float(df4)
df5 = convert_to_float(df5)
df7 = convert_to_float(df7)
df8 = convert_to_float(df8)

# Verificar el resultado en cada dataframe
print(df3.dtypes)
print(df4.dtypes)
print(df5.dtypes)
print(df7.dtypes)
print(df8.dtypes)
```

```python
# Para convertir todos los datos en enteros, primeramente se habian convertido a float

# Función para verificar si hay valores decimales en una columna
def tiene_decimales(columna):
    return any(valor % 1 != 0 for valor in columna.dropna())

# Convertir a tipo int solo si no hay valores decimales y no hay valores nulos en la columna
def convertir_a_int(df):
    for columna in df.select_dtypes(include=['float']).columns:
        if not tiene_decimales(df[columna]) and df[columna].notnull().all():
            df[columna] = df[columna].astype(int)
    return df

# Aplicar la conversión a tipo int para cada dataframe
df3 = convertir_a_int(df3)
df4 = convertir_a_int(df4)
df5 = convertir_a_int(df5)
df7 = convertir_a_int(df7)
df8 = convertir_a_int(df8)

# Verificar el resultado en cada dataframe
print(df3.dtypes)
print(df4.dtypes)
print(df5.dtypes)
print(df7.dtypes)
print(df8.dtypes)

```
